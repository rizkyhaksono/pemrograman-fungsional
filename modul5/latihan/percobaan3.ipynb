```python
__author__ = "rizkyhaksono"
__copyright__ = "Copyright 2023, Malang"

import matplotlib.pyplot as plt
import numpy as np

grafik1 = np.array([1, 4, 8, 3])
grafik2 = np.array([6, 2, 7, 11])
grafik3 = np.array([9, 2, 4, 7])

plt.plot(grafik1, label="Grafik A", marker="H", color="pink", ls="dotted")
plt.plot(grafik2, label="Grafik B", color="yellow")
plt.plot(grafik3, label="Grafik C", marker="*", ls="dashed", color="blue")

plt.title("Plot Tiga Garis")
plt.xlabel("Nilai x")
plt.ylabel("Nilai y")

plt.legend()
plt.show()
```
